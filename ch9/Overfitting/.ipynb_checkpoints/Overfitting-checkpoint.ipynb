```python
# パッケージのインポート
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
```

```python
# データセットの準備
imdb = keras.datasets.imdb
(train_data, train_labels), (test_data, test_labels) = imdb.load_data(num_words=10000)
```

```python
# One Hot　encoding
def multi_hot_sequences(sequences, dimension):
    results = np.zeros((len(sequences), dimension))
    for i, word_indices in enumerate(sequences):
        results[i, word_indices] = 1.0
    return results

# 訓練データと評価データをOne Hot encoding
train_data = multi_hot_sequences(train_data, dimension=10000)
test_data = multi_hot_sequences(test_data, dimension=10000)
```

```python
# 先頭の訓練データの確認
train_data[0]
```

```python
# 先頭の訓練データのプロット
plt.plot(train_data[0])
```

```python
# 容量が普通なモデル構造の作成
baseline_model = keras.Sequential([
    keras.layers.Dense(16, activation=tf.nn.relu, input_shape=(10000,)),
    keras.layers.Dense(16, activation=tf.nn.relu),
    keras.layers.Dense(1, activation=tf.nn.sigmoid)
])

# モデルのコンパイル
baseline_model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy', 'binary_crossentropy'])

# モデルの概要の表示
baseline_model.summary()

# 学習
baseline_history = baseline_model.fit(
    train_data,
    train_labels,
    epochs=20,
    batch_size=512,
    validation_data=(test_data, test_labels),
    verbose=2)
```

```python
# 容量が小さなモデル構造の作成
smaller_model = keras.Sequential([
    keras.layers.Dense(4, activation=tf.nn.relu, input_shape=(10000,)),
    keras.layers.Dense(4, activation=tf.nn.relu),
    keras.layers.Dense(1, activation=tf.nn.sigmoid)
])

# モデルのコンパイル
smaller_model.compile(optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy', 'binary_crossentropy'])

# モデルの概要の表示
smaller_model.summary()

# 学習
smaller_history = smaller_model.fit(
    train_data,
    train_labels,
    epochs=20,
    batch_size=512,
    validation_data=(test_data, test_labels),
    verbose=2)
```

```python
# 容量が大きなモデル構造の作成
bigger_model = keras.models.Sequential([
    keras.layers.Dense(512, activation=tf.nn.relu, input_shape=(10000,)),
    keras.layers.Dense(512, activation=tf.nn.relu),
    keras.layers.Dense(1, activation=tf.nn.sigmoid)
])

# モデルのコンパイル
bigger_model.compile(optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy','binary_crossentropy'])

# モデルの概要の表示
bigger_model.summary()

# 学習
bigger_history = bigger_model.fit(
    train_data, train_labels,
    epochs=20,
    batch_size=512,
    validation_data=(test_data, test_labels),
    verbose=2)
```

```python
# グラフ表示
def plot_history(histories, key='binary_crossentropy'):
    plt.figure(figsize=(16,10))
    for name, history in histories:
        val = plt.plot(history.epoch, history.history['val_'+key],
            '--', label=name.title()+' Val')
        plt.plot(history.epoch, history.history[key], color=val[0].get_color(),
            label=name.title()+' Train')
    plt.xlabel('Epochs')
    plt.ylabel(key.replace('_',' ').title())
    plt.legend()
    plt.xlim([0,max(history.epoch)])

# グラフ表示の実行
plot_history([
    ('bigger', bigger_history),
    ('baseline', baseline_history),
    ('smaller', smaller_history)])
```

```python
# 正則化モデル構造の作成
l2_model = keras.models.Sequential([
    keras.layers.Dense(16, 
        kernel_regularizer=keras.regularizers.l2(0.001),
        activation=tf.nn.relu, input_shape=(10000,)),
    keras.layers.Dense(16, 
        kernel_regularizer=keras.regularizers.l2(0.001),
        activation=tf.nn.relu),
    keras.layers.Dense(1, activation=tf.nn.sigmoid)
])

# モデルのコンパイル
l2_model.compile(optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy', 'binary_crossentropy'])

# 学習
l2_model_history = l2_model.fit(
    train_data, train_labels,
    epochs=20,
    batch_size=512,
    validation_data=(test_data, test_labels),
    verbose=2)

# グラフの表示
plot_history([
    ('baseline', baseline_history),
    ('l2', l2_model_history)])
```

```python
# ドロップモデル構造の作成
dpt_model = keras.models.Sequential([
    keras.layers.Dense(16, activation=tf.nn.relu, input_shape=(10000,)),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(16, activation=tf.nn.relu),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1, activation=tf.nn.sigmoid)
])

# モデルのコンパイル
dpt_model.compile(optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy','binary_crossentropy'])

# 学習
dpt_model_history = dpt_model.fit(
    train_data, train_labels,
    epochs=20,
    batch_size=512,
    validation_data=(test_data, test_labels),
    verbose=2)

# グラフの表示
plot_history([
    ('baseline', baseline_history),
    ('dropout', dpt_model_history)])
```

```python
# ドロップモデル構造の作成
dpt_model = keras.models.Sequential([
    keras.layers.Dense(8, activation=tf.nn.relu, input_shape=(10000,)),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(8, activation=tf.nn.relu),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1, activation=tf.nn.sigmoid)
])

# モデルのコンパイル
dpt_model.compile(optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy','binary_crossentropy'])

# 学習
dpt_model_history = dpt_model.fit(
    train_data, train_labels,
    epochs=20,
    batch_size=512,
    validation_data=(test_data, test_labels),
    verbose=2)

# グラフの表示
plot_history([
    ('baseline', baseline_history),
    ('dropout', dpt_model_history)])
```
