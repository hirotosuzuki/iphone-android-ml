```python
import turicreate as tc

# データセットの読み込み
styles = tc.image_analysis.load_images('./style/')
content = tc.image_analysis.load_images('./content/')

# データセットの確認
styles.explore()
```

```python
# 学習
model = tc.style_transfer.create(styles, content)

# Turi Createモデルの保存
model.save('./StyleTransfer.model')

# Core MLモデルの保存
model.export_coreml('./StyleTransfer.mlmodel')
```

```python
# 画風変換
stylize_result = model.stylize(content[0:1])
stylize_result.explore()
```
